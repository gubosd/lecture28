# 선형회귀

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

- 어떤 문제를 해결하려고 하는가?
> sepal length 값으로 sepal width 값을 예측하고 싶다

```python
X = iris.data[:, 0] # sepal length
y = iris.data[:, 1] # sepal width (목표값, 타겟값)
display(X, y)
```

```python
# X 값은 2차원 즉, 표형태의 데이터여야만 한다
X = X.reshape(-1, 1) # X.reshape(150,1)
X.shape
```

```python
model.fit(X, y)
```

```python
model.coef_ # 기울기
```

```python
model.intercept_ # 절편
```

```python
plt.scatter(iris.data[:,0], iris.data[:,1])
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
w = model.coef_
b = model.intercept_
```

```python
xval = np.array([4,8])
yval = w*xval + b
yval
```

```python
plt.scatter(iris.data[:,0], iris.data[:,1], alpha=0.5)
plt.xlabel('sepal length')
plt.ylabel('sepal width')

plt.plot(xval, yval, 'r-')
```

```python
model.predict([[5]])
```

```python
model.predict([[5],[6]])
```

```python
model.predict([[10]])
```

- 연습문제
> petal length 값으로 petal width 값을 선형회귀로 분석해 보시오 (직선 그리기)

```python
X = iris.data[:, 2]
X = X.reshape(-1, 1)
y = iris.data[:, 3]
X.shape, y.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
w = model.coef_
b = model.intercept_
w, b
```

```python
xval = np.array([0, 7])
yval = w*xval + b

plt.scatter(iris.data[:,2], iris.data[:,3], alpha=0.5)
plt.xlabel('petal length')
plt.ylabel('petal width')

plt.plot(xval, yval, 'r-')
```

```python
model.predict([[3]])
```

### 입력 속성이 여러개인 선형회귀

- 해결하려는 문제는?
> sepal length, sepal width 값을 가지고 setosa 와 versicolor 를 구분해 보자

```python
X = iris.data[:100, :2]
y = iris.target[:100] # 0: setosa, 1: versicolor
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.colorbar()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
w = model.coef_[0]
w
```

```python
b = model.intercept_
b
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.colorbar()
plt.xlabel('sepal length')
plt.ylabel('sepal width')

xval = np.array([4, 7.5])
yval = -(w[0]*xval + b)/w[1]
plt.plot(xval, yval)
```
